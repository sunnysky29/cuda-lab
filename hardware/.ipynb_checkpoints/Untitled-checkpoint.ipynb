## GPU通信问题

Kepler 架构中最值得一提的是 GPUDirect 技术，可以绕过 CPU/System Memory，完成与本机其他 GPU 或者其他机器 GPU 的直接数据交换。
如今，Bypass CPU/OS 已经是最重要加速手段之一。
